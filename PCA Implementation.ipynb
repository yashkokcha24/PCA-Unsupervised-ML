## Importing Dependencies

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
## Load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset= load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df= pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
print(df)
```

```python
df.head()
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data= scaler.transform(df)
```

```python
scaled_data
```

```python
## Applying PCA Algorithm
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)  ## N-components in we are having only 2 feature out of all the feature
```

```python
data_pca= pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_  ## explained_variance in we are able to get variance with 2 feature 
```

```python
## Let's suppose if we are not having 'n_component=2' that we will not be able to get variance properly
pca= PCA()
```

```python
pca.fit_transform(scaled_data)
```

```python
pca.explained_variance_   ## In this we are having variance with respect to more than 2 feature
```

```python
## Ploting the PCA (N_component= 2)
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
## For N_component= 3
pca=PCA(n_components=3)
```

```python
data_pca2= pca.fit_transform(scaled_data)
```
